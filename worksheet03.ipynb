```python
import pandas as pd

# Assuming df is the Titanic dataset loaded into a DataFrame

# Step 1: Add a new column 'Fare per Year' calculated as Fare / Age
df['Fare per Year'] = df['Fare'] / df['Age']

# Step 2: Subset rows where 'Fare per Year' is greater than 5
high_fare_age = df[df['Fare per Year'] > 5]

# Step 3: Sort 'high_fare_age' by 'Fare per Year' in descending order
high_fare_age_srt = high_fare_age.sort_values(by='Fare per Year', ascending=False)

# Step 4: Select only the 'Name' and 'Fare per Year' columns and save the result as 'result'
result = high_fare_age_srt[['Name', 'Fare per Year']]

# Print the result to view
result.head(100)
```

```python
import pandas as pd

# Assuming df is the Titanic dataset loaded into a DataFrame

# Step 1: Add a new column 'Fare per Class' calculated as Fare / Pclass
df['Fare per Class'] = df['Fare'] / df['Pclass']

# Step 2: Subset rows where passenger is male (Sex is "male") and adult (Age >= 18)
adult_males = df[(df['Sex'] == 'male') & (df['Age'] >= 18)]

# Step 3: Sort adult males by 'Fare per Class' in descending order
adult_males_srt = adult_males.sort_values(by='Fare per Class', ascending=False)

# Step 4: Select only 'Name', 'Age', and 'Fare per Class' columns and save the result as 'result'
result = adult_males_srt[['Name', 'Age', 'Fare per Class']]

# Print the result to view
result.head(100)
```

```python
import pandas as pd

# Assuming df is the Titanic dataset loaded into a DataFrame

# Step 1: Calculate the total Fare paid across all passengers
total_fare = df['Fare'].sum()

# Step 2: Subset for passengers in first class (Pclass is 1) and calculate their total fare
first_class_fare = df[df['Pclass'] == 1]['Fare'].sum()

# Step 3: Subset for passengers in second class (Pclass is 2) and calculate their total fare
second_class_fare = df[df['Pclass'] == 2]['Fare'].sum()

# Subset for passengers in third class (Pclass is 3) and calculate their total fare
third_class_fare = df[df['Pclass'] == 3]['Fare'].sum()

# Step 4: Combine the fare totals from first, second, and third classes into a list
fare_totals_by_class = [first_class_fare, second_class_fare, third_class_fare]

# Step 5: Divide the totals for each class by the overall total fare to get the proportion of fare revenue by class
proportions_by_class = [fare / total_fare for fare in fare_totals_by_class]

# Print the results
print(f"Total Fare: {total_fare}")
print(f"First Class Fare: {first_class_fare}")
print(f"Second Class Fare: {second_class_fare}")
print(f"Third Class Fare: {third_class_fare}")
print(f"Fare Totals by Class: {fare_totals_by_class}")
print(f"Proportions by Class: {proportions_by_class}")
```

```python
import pandas as pd

# Assuming df is the Titanic dataset loaded into a DataFrame

# Step 1: Create a new column 'age group' that categorizes passengers
def categorize_age(age):
    if age < 18:
        return 'child'
    elif 18 <= age < 65:
        return 'adult'
    else:
        return 'senior'

df['age group'] = df['Age'].apply(categorize_age)

# Step 2: Calculate the total number of passengers on the Titanic
total_passengers = df.shape[0]

# Step 3: Count the number of passengers in each age group
age_group_counts = df['age group'].value_counts()

# Step 4: Divide the count of each age group by the total number of passengers
age_group_proportions = age_group_counts / total_passengers

# Step 5: Display the proportion as a percentage
age_group_percentage = age_group_proportions * 100

# Print the results
print(f"Total Number of Passengers: {total_passengers}")
print(f"Age Group Counts:\n{age_group_counts}")
print(f"Proportion of Passengers in Each Age Group:\n{age_group_proportions}")
print(f"Proportion of Passengers in Each Age Group (Percentage):\n{age_group_percentage}")
```
