Complete the following Tasks:
1. Initialize an empty array with size 2X2.
2. Initialize an all one array with size 4X2.
3. Return a new array of given shape and type, filled with fill value.{Hint: np.full}
4. Return a new array of zeros with same shape and type as a given array.{Hint: np.zeros like}
5. Return a new array of ones with same shape and type as a given array.{Hint: np.ones like}
6. For an existing list new_list = [1,2,3,4] convert to an numpy array.{Hint: np.array()}

```python
import numpy as np

# Task 1: Initialize an empty array with size 2X2
empty_array = np.empty((2, 2))
print("Empty Array (2x2):")
print(empty_array)

# Task 2: Initialize an all ones array with size 4X2
ones_array = np.ones((4, 2))
print("\nAll Ones Array (4x2):")
print(ones_array)

# Task 3: Return a new array of given shape and type, filled with a specified fill value
fill_value_array = np.full((3, 3), 1)  # Example shape (3,3) and fill value 7
print("\nArray filled with 1 (3x3):")
print(fill_value_array)

# Task 4: Return a new array of zeros with same shape and type as a given array
example_array = np.array([[1, 2], [3, 4]])
zeros_like_array = np.zeros_like(example_array)
print("\nZeros Array with Same Shape as example_array:")
print(zeros_like_array)

# Task 5: Return a new array of ones with same shape and type as a given array
ones_like_array = np.ones_like(example_array)
print("\nOnes Array with Same Shape as example_array:")
print(ones_like_array)

# Task 6: Convert an existing list to a NumPy array
new_list = [1, 2, 3, 4]
numpy_array = np.array(new_list)
print("\nConverted NumPy Array from List:")
print(numpy_array)
```

```python
import numpy as np

#task-1
array_range = np.arange(10, 50)
print("Array with value from 10-49: ")
print(array_range)

#task-2
matrix_3x3 = np.arange(9).reshape((3,3))
print("\n Matrix with value from 3-8: ")
print(matrix_3x3)

#task-3
identity_matrix = np.eye(3)     #it will make a matrix of three columns and rows
print("\n 3x3 indentity matrix: ")
print(identity_matrix)

#task-4
random_array = np.random.random(30)
mean_value = random_array.mean()
print("\n random array of size 30: ", random_array)
print("\n mean: ", mean_value)

#task-5
random_10x10 = np.random.random((10,10))
min_value = random_10x10.min()
max_value = random_10x10.max()
print("\n Random value: ", random_10x10)
print("\n max value: ", max_value)
print("\n min value: ", min_value)

#task-6
zero_array = np.zeros(10)
zero_array[4] = 4
print("\n Zero array with fifth element replaced: ", zero_array)

#task-7
arr = [1, 2, 0, 0, 4, 0]
reversed_arr = arr[::-1]
print("\n Reversed_arr: ", reversed_arr)


#task-8
border_array = np.ones((5, 5))
border_array[1:-1, 1:-1] = 0
print("\n2D Array with 1 on border and 0 inside: ")
print(border_array)

#task-9
checkerboard = np.zeros((8, 8))
checkerboard[1::2, ::2] = 1  # Fill odd rows
checkerboard[::2, 1::2] = 1  # Fill even rows
print("\n8x8 Checkerboard Pattern:")
print(checkerboard)

```

• Solve the following system of Linear equation using Inverse Methods.

2x − 3y + z = −1
x − y + 2z = −3
3x + y − z = 9

{Hint: First use Numpy array to represent the equation in Matrix form. Then Solve for: AX = B}
• Now: solve the above equation using np.linalg.inv function.{Explore more about ”linalg” function
of Numpy}

```python
import numpy as np

# Given arrays
A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])

# Task 1: Prove A.A^-1 = I
A_inv = np.linalg.inv(A)
identity_matrix = np.dot(A, A_inv)
print("Task 1 - Prove A.A^-1 = I:")
print(identity_matrix)

# Task 2: Prove AB ≠ BA
AB = np.dot(A, B)
BA = np.dot(B, A)
print("\nTask 2 - Prove AB ≠ BA:")
print("AB =")
print(AB)
print("BA =")
print(BA)

# Task 3: Prove (AB)^T = B^T A^T
AB_transpose = np.transpose(AB)
B_transpose_A_transpose = np.dot(np.transpose(B), np.transpose(A))
print("\nTask 3 - Prove (AB)^T = B^T A^T:")
print("Transpose of AB:")
print(AB_transpose)
print("B^T A^T:")
print(B_transpose_A_transpose)

# Task 4: Solve the system of linear equations using inverse method
A_matrix = np.array([[2, -3, 1], [1, -1, 2], [3, 1, -1]])  # Coefficient matrix A
B_matrix = np.array([-1, -3, 9])  # Right-hand side matrix B

# Inverse method
A_inv = np.linalg.inv(A_matrix)
X = np.dot(A_inv, B_matrix)

print("\nTask 4 - Solve the system of equations using Inverse Methods:")
print("Solution for x, y, z:")
print(X)
```
